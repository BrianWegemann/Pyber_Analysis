```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
type(np.array([1,2,3,4]))
```

```python
x_axis = np.arange(0,6,0.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

```python
plt.plot(x_axis, sin)
plt.plot(x_axis, cos)
```

```python
ex_p = [np.exp(x) for x in x_axis]
```

```python
a = []
for i in x_axis:
    a.append(np.exp(i))
a
```

```python
plt.plot(x_axis, ex_p)
plt.xlabel('What this one is')
plt.ylabel('Exponential Function')
plt.title("Pick Title")
```
